Since all these datasets are of the same format, we can append only the data pertaining to India from each dataset to this dataframe using an iterator.

```python
external_data_files = os.listdir("../data/external/")
```

```python
for file in external_data_files:
    df = pd.read_csv(f"../data/external/{file}", skiprows=4)
    print(file)
    india_row = df[df["Country Name"] == "India"]
    combined_economic_indicators = pd.concat([combined_economic_indicators, india_row], ignore_index=True)
    # combined_economic_indicators = pd.concat([combined_economic_indicators, df], ignore_index=True)
```

# Wrangling and Cleaning Candidate List

```python
cand_list_2024 = pd.read_csv("../data/raw/candidates-2024.csv")
```

```python
cand_list_2024.tail()
```

```python
cand_list_2024.drop('Photo Link', axis=1,inplace=True)
```

```python
cand_list_2024
```

```python
cand_list_2024.info()
```

Let us change the date format of both the date columns.

```python
import re

# Define a function to clean the 'th', 'rd', 'st', and 'nd' from the date
def clean_date(date_str):
    return re.sub(r'(\d+)(st|nd|rd|th)', r'\1', date_str)

# Apply the function to clean the 'Election Date' column
cand_list_2024['Application Date'] = cand_list_2024['Application Date'].apply(clean_date)

# Convert the cleaned 'Election Date' column to datetime
cand_list_2024['Application Date'] = pd.to_datetime(cand_list_2024['Application Date'], format='%d %B, %Y')

# View the result
```

```python
cand_list_2024
```

```python
cand_list_2024['Polling Date'] = pd.to_datetime(cand_list_2024['Polling Date'])
```

```python
cand_list_2024
```

Dropping other irrelavant columns:

```python
cand_list_2024.drop(["Address","Father/Husband's Name"], axis=1,inplace=True)
```

```python
cand_list_2024
```

```python
# Count rows where 'Application Status' is not 'Accepted'
print(cand_list_2024[cand_list_2024['Application Status'] != 'Accepted'].shape[0])
```

Since there are no entries with Application Status other than Accepted we will drop this column too. Along with that let us also drop the _id column.

```python
cand_list_2024.drop(["_id","Application Status"], axis=1,inplace=True)
```

```python
cand_list_2024
```

The column Application isn't very important so let's drop this column too.

```python
cand_list_2024.drop(["Application Date"], axis=1,inplace=True)
```
