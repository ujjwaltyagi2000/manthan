```python
import re

# Define a function to clean the 'th', 'rd', 'st', and 'nd' from the date
def clean_date(date_str):
    return re.sub(r'(\d+)(st|nd|rd|th)', r'\1', date_str)

# Apply the function to clean the 'Election Date' column
cand_list_2024['Application Date'] = cand_list_2024['Application Date'].apply(clean_date)

# Convert the cleaned 'Election Date' column to datetime
cand_list_2024['Application Date'] = pd.to_datetime(cand_list_2024['Application Date'], format='%d %B, %Y')

# View the result
```

# Wrangling and cleaning candidate background datasets:

```python
# Load the dataset
cand_bg_2024 = pd.read_csv("../data/raw/candidate_background_2024.csv")
```

```python
cand_bg_2024.columns
```

```python
cand_bg_2024
```

We first need to change the units of Total Assets and Liabilities and convert them to integers 

```python
def clean_currency(value):
    if pd.isnull(value) or value.strip() == '':  # Handle NaN or empty values
        return 0
    
    # Split the string at the '~' symbol and take the part before it
    value = value.split('~')[0]
    
    # Remove non-numeric characters
    cleaned_value = re.sub(r'[^\d]', '', value)
    
    # If cleaned_value is empty after removing non-numeric characters, return 0
    if cleaned_value == '':
        return 0
    
    return int(cleaned_value)
```

```python

# Apply the function to the 'Total Assets' and 'Liabilities' columns
cand_bg_2024['Total Assets'] = cand_bg_2024['Total Assets'].apply(clean_currency)
cand_bg_2024['Liabilities'] = cand_bg_2024['Liabilities'].apply(clean_currency)
```

```python
cand_bg_2024
```

```python
cand_bg_2024.columns
```

```python
cand_bg_2024.rename(columns={
    'Candidate': 'Candidate Name',
}, inplace=True)
cand_bg_2024 = cand_bg_2024[['Candidate Name', 'Party', 'Age', 'State', 'Criminal Cases', 'Education', 'Total Assets', 'Liabilities']]
cand_bg_2024
```

```python
cand_bg_2024.to_csv("../data/processed/candidate_background_2024.csv", index=False)
```
