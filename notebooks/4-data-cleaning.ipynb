```python
import pandas as pd
import numpy as np
from fuzzywuzzy import fuzz
from tqdm import tqdm

# Load the datasets
# cand_list_2024 = pd.read_csv('cand_list_2024.csv')
# cands = pd.read_csv('cands.csv')

# Print column names to verify
print("Columns in cand_list_2024:", cand_list_2024.columns)
print("Columns in cands:", cands.columns)

# Function to clean and standardize names
def clean_name(name):
    name = str(name).lower()
    name = name.replace('.', ' ').replace(',', ' ')
    name = ' '.join(name.split())  # Remove extra spaces
    return name

# Identify the correct column names for candidate names
candidate_col_2024 = 'Candidate' if 'Candidate' in cand_list_2024.columns else input("Enter the column name for candidate names in cand_list_2024: ")
candidate_col_cands = 'Candidate Name' if 'Candidate Name' in cands.columns else input("Enter the column name for candidate names in cands: ")

# Apply name cleaning to both datasets
cand_list_2024['Clean_Candidate'] = cand_list_2024[candidate_col_2024].apply(clean_name)
cands['Clean_Candidate'] = cands[candidate_col_cands].apply(clean_name)

# Function to find the best match
def find_best_match(name, state, constituency, candidates):
    best_match = None
    best_score = 0
    for _, row in candidates.iterrows():
        if row['State'] == state and row['Constituency'] == constituency:
            score = fuzz.token_set_ratio(name, row['Clean_Candidate'])
            if score > best_score:
                best_score = score
                best_match = row
    return best_match if best_score > 80 else None

# Create a dictionary to store matches
matches = {}

# Find matches for each candidate
for _, row in tqdm(cand_list_2024.iterrows(), total=len(cand_list_2024)):
    key = (row['Clean_Candidate'], row['State'], row['Constituency'])
    if key not in matches:
        match = find_best_match(row['Clean_Candidate'], row['State'], row['Constituency'], cands)
        matches[key] = match

# Function to get age and gender from matches
def get_age_gender(clean_name, state, constituency):
    match = matches.get((clean_name, state, constituency))
    return pd.Series({'Age': match['Age'] if match is not None else np.nan,
                      'Gender': match['Gender'] if match is not None else np.nan})

# Add Age and Gender columns to cand_list_2024
cand_list_2024[['Age', 'Gender']] = cand_list_2024.apply(
    lambda row: get_age_gender(row['Clean_Candidate'], row['State'], row['Constituency']), axis=1
)

# Remove the temporary 'Clean_Candidate' column
cand_list_2024 = cand_list_2024.drop('Clean_Candidate', axis=1)

# Save the merged dataset
cand_list_2024.to_csv('merged_lok_sabha_2024_data.csv', index=False)

print("Merged data saved to 'merged_lok_sabha_2024_data.csv'")
```
