## **1. Shift in Economic Indicators**

```python
# imports
from plotly.subplots import make_subplots

import plotly.graph_objects as go
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# Filter the row corresponding to 'Access to electricity (% of population)'
electricity_df = comb_eco_df[comb_eco_df['indicator_name'] == 'Access to electricity (% of population)']

# Melt the DataFrame to convert the wide format into a long format
electricity_df_melted = electricity_df.melt(id_vars=['indicator_name'], var_name='Year', value_name='Access to electricity')

# Drop rows with missing values
electricity_df_melted.dropna(subset=['Access to electricity'], inplace=True)

# Convert the Year column to numeric (just in case)
electricity_df_melted['Year'] = pd.to_numeric(electricity_df_melted['Year'])

# Plotting with seaborn
plt.figure(figsize=(10, 6))
sns.lineplot(data=electricity_df_melted, x='Year', y='Access to electricity', marker='o')

# Highlight markers for the years 2014, 2019, and 2024
highlight_years = [2014, 2019, 2024]
highlight_data = electricity_df_melted[electricity_df_melted['Year'].isin(highlight_years)]

# Plot the markers for the highlighted years
plt.scatter(highlight_data['Year'], highlight_data['Access to electricity'], color='red', s=100, zorder=5, label="NDA Government (2014,2019)")

# Customizing the plot
plt.title('Access to Electricity (% of Population) Over Time')
plt.xlabel('Year')
plt.ylabel('Access to Electricity (% of Population)')
plt.grid(True)
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()

# Show the plot
plt.show()
```

```python
indicators = ['Access to electricity (% of population)',
       'Foreign direct investment, net inflows (% of GDP)',
       'GDP growth (annual %)', 'GDP per capita (current US$)',
       'GDP (current US$)', 'Central government debt, total (% of GDP)',
       'Inflation, consumer prices (annual %)', 'Net migration',
       'Poverty headcount ratio at $2.15 a day (2017 PPP) (% of population)',
       'Statistical performance indicators (SPI): Overall score (scale 0-100)',
       'Unemployment, total (% of total labor force) (modeled ILO estimate)',
       'Proportion of seats held by women in national parliaments (%)']
```

```python
# Melt the DataFrame to long format
comb_eco_melted = comb_eco_df.melt(id_vars=['indicator_name'], var_name='Year', value_name='Value')

# Convert the Year column to numeric and drop rows with missing values
comb_eco_melted['Year'] = pd.to_numeric(comb_eco_melted['Year'], errors='coerce')
comb_eco_melted.dropna(subset=['Value'], inplace=True)

# Filter for the relevant indicators
filtered_df = comb_eco_melted[comb_eco_melted['indicator_name'].isin(indicators)]

# Set up the matplotlib figure and axes for subplots
fig, axes = plt.subplots(len(indicators), 1, figsize=(10, len(indicators) * 3))

# Loop through each indicator and plot on its respective subplot
for i, indicator in enumerate(indicators):
    indicator_df = filtered_df[filtered_df['indicator_name'] == indicator]
    sns.lineplot(data=indicator_df, x='Year', y='Value', ax=axes[i], marker='o')
    axes[i].set_title(indicator)
    axes[i].set_xlabel('Year')
    axes[i].set_ylabel('Value')
    axes[i].grid(True)

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```

Here are some key insights from the provided data across different indicators:

1. **Access to Electricity**:
   - From the mid-1980s, access to electricity started rising significantly, with major growth visible from 1990 onwards.
   - By 2020, almost universal access to electricity was achieved, with over 99% of the population having access.
   
2. **Foreign Direct Investment (FDI) Inflows**:
   - FDI inflows remained relatively low and unstable until the 1990s, with occasional negative values, likely indicating capital outflows.
   - A significant rise in FDI started in the mid-1990s, with rapid growth continuing into the 2000s, peaking at around 3.6% of GDP in 2008.
   - Afterward, there are fluctuations but still significant FDI inflows, reflecting India's growing integration into the global economy.

3. **GDP Growth**:
   - The data shows periods of rapid growth interspersed with economic slowdowns. Notable dips occur in 1965, 1979, 1991, and 2020 (due to external shocks or global crises).
   - Noteworthy growth periods include the mid-1980s, late 1990s, and mid-2000s, aligning with major economic reforms.
   - The 2020 dip corresponds to the global impact of the COVID-19 pandemic, with a significant rebound in 2021.

4. **GDP Per Capita**:
   - GDP per capita rose steadily, especially from the 1990s onward, reflecting India's economic expansion.
   - The biggest jump happened in the early 2000s and continued through the 2010s, showing improved living standards. However, there’s a slight decline post-2020, likely due to the pandemic.

5. **Government Debt as % of GDP**:
   - Central government debt gradually decreased from the mid-2000s to 2019, indicating better fiscal management.
   - In the 2000s, the government debt was as high as 62%, but it dropped to around 46% by the late 2010s.
   
6. **Poverty Reduction**:
   - There is a clear decline in the poverty headcount ratio from 63.5% in the early years to 13.2% in recent years, showcasing significant progress in poverty alleviation.
   
7. **Unemployment**:
   - The unemployment rate fluctuated slightly but remained fairly high throughout, peaking in the 2000s and gradually declining post-2010.
   - There was a notable dip in unemployment from 2020, possibly due to policy changes or economic recovery efforts post-COVID.

8. **Proportion of Seats Held by Women in National Parliament**:
   - Representation of women in parliament has grown steadily from about 7% in the early 2000s to over 15% in recent years. Although progress is evident, there’s still room for more gender inclusivity.

These indicators provide insights into India's economic, social, and political development, showcasing strides in infrastructure, foreign investment, economic growth, poverty reduction, and gender representation in parliament. However, challenges like consistent unemployment and fiscal management remain key areas to address.

From the age distribution graph and average age data, we observe several trends for the years 2019 and 2024. Focusing first on the **BJP** and **INC**, we can see that the **BJP** has a slight increase in average age from **54.65** in 2019 to **56.59** in 2024, reflecting a preference for slightly older candidates in 2024. This trend is further supported by the distribution, where the age group around 60 appears more prominent in 2024. In contrast, **INC** shows a slight decrease in average age, from **55.08** in 2019 to **54.25** in 2024, indicating a shift towards younger candidates, though the difference is marginal.

Looking at other major parties, **AAP** shows a notable increase in average age from **48.56** in 2019 to **51.41** in 2024, suggesting that the party is leaning towards slightly more mature candidates. Similarly, **AITC** also sees a rise, from **52.05** in 2019 to **55.71** in 2024, while **SP** demonstrates a decrease from **54.53** in 2019 to **53.07** in 2024, suggesting a move towards younger candidates. **CPI(M)** and **DMK** maintain relatively stable average ages, with small fluctuations, while **Shiv Sena** saw no candidates in 2019 but has an average age of **54.93** in 2024. 

When looking at state-wise averages, the data reveals that some states have distinct age preferences for candidates, such as **Andaman and Nicobar Islands**, where the average age jumps from **44.25** in 2019 to **63.00** in 2024, and **Chandigarh**, with a significant shift from **71.33** in 2019 to **59.00** in 2024. Additionally, **Haryana** shows a significant increase in average age, from **53.46** in 2019 to **61.60** in 2024. This indicates possible regional variations in candidate age preferences across different states.

In summary, the age of candidates across the major parties has shown varying trends, with some parties favoring slightly older candidates in 2024, while others prefer younger representatives. The insights drawn from these trends can help us understand party strategies in candidate selection and the potential influence of candidate age on electoral outcomes.

```python
# Function to create a subplot with pie charts for both years
def create_pie_chart_comparison(data, party):
    # Filter data for the specified party and each year
    data_2019 = data[(data['party'] == party) & (data['year'] == 2019)]['education'].value_counts(normalize=True) * 100
    data_2024 = data[(data['party'] == party) & (data['year'] == 2024)]['education'].value_counts(normalize=True) * 100
    
    # Create subplots for the two years
    fig = make_subplots(
        rows=1, cols=2, 
        specs=[[{'type': 'domain'}, {'type': 'domain'}]], 
        subplot_titles=[f'{party} (2019)', f'{party} (2024)']
    )
    
    # Add pie chart for 2019
    fig.add_trace(
        go.Pie(
            labels=data_2019.index,
            values=data_2019.values,
            name='2019',
            hole=0.3,
            textinfo='percent+label'
        ),
        row=1, col=1
    )
    
    # Add pie chart for 2024
    fig.add_trace(
        go.Pie(
            labels=data_2024.index,
            values=data_2024.values,
            name='2024',
            hole=0.3,
            textinfo='percent+label'
        ),
        row=1, col=2
    )
    
    # Update the layout
    fig.update_layout(
        title_text=f'Education Background Distribution - {party}',
        title_x=0.5,  # Center the title
        showlegend=False,  # Hide legends for better clarity
    )
    
    # Show the figure
    fig.show()

# Generate the pie chart comparison for all parties
parties = top_parties_df['party'].unique()
for party in parties:
    create_pie_chart_comparison(top_parties_df, party)
```

```python
# Function to create a comparison of matriculation-passed candidates for both years
def create_matriculation_comparison(data):
    parties = data['party'].unique()
    
    # Initialize lists to store counts for each party
    matriculation_passed_2019 = []
    matriculation_passed_2024 = []
    party_names = []

    # Loop through each party to calculate matriculation-passed candidates
    for party in parties:
        party_data_2019 = data[(data['party'] == party) & (data['year'] == 2019)]
        party_data_2024 = data[(data['party'] == party) & (data['year'] == 2024)]

        # Define levels of education indicating matriculation passed
        education_levels = [
            '10th Pass', '12th Pass', 'Graduate', 'Post Graduate', 
            'Doctorate', 'Professional Degree'
        ]

        # Calculate matriculation-passed counts
        passed_2019 = party_data_2019[party_data_2019['education'].isin(education_levels)].shape[0]
        passed_2024 = party_data_2024[party_data_2024['education'].isin(education_levels)].shape[0]

        # Append data to the lists
        matriculation_passed_2019.append(passed_2019)
        matriculation_passed_2024.append(passed_2024)
        party_names.append(party)

    # Create a subplot with bar charts for both years
    fig = go.Figure()

    # Add bar for 2019
    fig.add_trace(
        go.Bar(
            x=party_names,
            y=matriculation_passed_2019,
            name='2019',
            marker=dict(color='navy')
        )
    )

    # Add bar for 2024
    fig.add_trace(
        go.Bar(
            x=party_names,
            y=matriculation_passed_2024,
            name='2024',
            marker=dict(color='orange')
        )
    )

    # Update layout
    fig.update_layout(
        title_text='Matriculation-Passed Candidates by Party (2019 vs 2024)',
        title_x=0.5,  # Center the title
        barmode='group',  # Group bars side by side
        xaxis_title='Party',
        yaxis_title='Number of Matriculation-Passed Candidates',
        legend_title='Year',
        width=1200,  # Increase figure width
        height=800   # Increase figure height
    )

    # Show the figure
    fig.show()

# Generate the comparison chart
create_matriculation_comparison(top_parties_df)
```

```python
import pandas as pd

# Assuming `top_parties_df` is your DataFrame

def identify_outliers(df, column):
    """
    Identify outliers using the Interquartile Range (IQR) method.
    """
    Q1 = df[column].quantile(0.25)  # First quartile (25th percentile)
    Q3 = df[column].quantile(0.75)  # Third quartile (75th percentile)
    IQR = Q3 - Q1                  # Interquartile range
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

# Identify outliers in Net Worth and Criminal Cases for 2019 and 2024
outliers_net_worth_2019 = identify_outliers(data_2019, 'net_worth')
outliers_criminal_cases_2019 = identify_outliers(data_2019, 'criminal_cases')

outliers_net_worth_2024 = identify_outliers(data_2024, 'net_worth')
outliers_criminal_cases_2024 = identify_outliers(data_2024, 'criminal_cases')

# Combine the outliers for reporting
outliers_2019 = pd.concat([outliers_net_worth_2019, outliers_criminal_cases_2019]).drop_duplicates()
outliers_2024 = pd.concat([outliers_net_worth_2024, outliers_criminal_cases_2024]).drop_duplicates()

# Print outlier candidates
print("Outliers for 2019:")
print(outliers_2019[['candidate_name', 'party', 'net_worth', 'criminal_cases']])

print("\nOutliers for 2024:")
print(outliers_2024[['candidate_name', 'party', 'net_worth', 'criminal_cases']])
```
