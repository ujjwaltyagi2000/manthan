# Manthan: Lok Sabha Elections Analysis (2019 vs 2024)

## Project Overview

Manthan, meaning "churning" or "deep contemplation" in Sanskrit, is the name I chose for this data analytics project. My goal is to perform a comprehensive analysis of the Lok Sabha election results from 2019 and 2024, focusing on the performance shifts of the Bharatiya Janata Party (BJP) and the Indian National Congress (INC).

### Problem Statement

This project aims to analyze the factors contributing to the shift in electoral outcomes between the 2019 and 2024 Lok Sabha elections in India. I will identify key demographic, socio-economic, and political variables that may have influenced voting patterns and party performance across different constituencies.

Here are some key insights from the provided data across different indicators:

1. **Access to Electricity**:
   - From the mid-1980s, access to electricity started rising significantly, with major growth visible from 1990 onwards.
   - By 2020, almost universal access to electricity was achieved, with over 99% of the population having access.
   
2. **Foreign Direct Investment (FDI) Inflows**:
   - FDI inflows remained relatively low and unstable until the 1990s, with occasional negative values, likely indicating capital outflows.
   - A significant rise in FDI started in the mid-1990s, with rapid growth continuing into the 2000s, peaking at around 3.6% of GDP in 2008.
   - Afterward, there are fluctuations but still significant FDI inflows, reflecting India's growing integration into the global economy.

3. **GDP Growth**:
   - The data shows periods of rapid growth interspersed with economic slowdowns. Notable dips occur in 1965, 1979, 1991, and 2020 (due to external shocks or global crises).
   - Noteworthy growth periods include the mid-1980s, late 1990s, and mid-2000s, aligning with major economic reforms.
   - The 2020 dip corresponds to the global impact of the COVID-19 pandemic, with a significant rebound in 2021.

4. **GDP Per Capita**:
   - GDP per capita rose steadily, especially from the 1990s onward, reflecting India's economic expansion.
   - The biggest jump happened in the early 2000s and continued through the 2010s, showing improved living standards. However, thereâ€™s a slight decline post-2020, likely due to the pandemic.

5. **Government Debt as % of GDP**:
   - Central government debt gradually decreased from the mid-2000s to 2019, indicating better fiscal management.
   - In the 2000s, the government debt was as high as 62%, but it dropped to around 46% by the late 2010s.
   
6. **Poverty Reduction**:
   - There is a clear decline in the poverty headcount ratio from 63.5% in the early years to 13.2% in recent years, showcasing significant progress in poverty alleviation.
   
7. **Unemployment**:
   - The unemployment rate fluctuated slightly but remained fairly high throughout, peaking in the 2000s and gradually declining post-2010.
   - There was a notable dip in unemployment from 2020, possibly due to policy changes or economic recovery efforts post-COVID.

8. **Proportion of Seats Held by Women in National Parliament**:
   - Representation of women in parliament has grown steadily from about 7% in the early 2000s to over 15% in recent years. Although progress is evident, thereâ€™s still room for more gender inclusivity.

These indicators provide insights into India's economic, social, and political development, showcasing strides in infrastructure, foreign investment, economic growth, poverty reduction, and gender representation in parliament. However, challenges like consistent unemployment and fiscal management remain key areas to address.

```python
import plotly.express as px

# Function to create a pie chart for a specific party
def create_pie_chart(data, party):
    party_data = data[data['party'] == party]['education'].value_counts(normalize=True) * 100
    labels = party_data.index
    sizes = party_data.values
    
    # Create the pie chart
    fig = px.pie(
        values=sizes,
        names=labels,
        title=f'Education Background Distribution - {party}',
        hole=0.3  # To make it a donut chart, adjust the hole size
    )
    fig.update_traces(textinfo='percent+label')
    fig.show()

# Generate pie charts for all parties
parties = top_parties_df['party'].unique()
for party in parties:
    create_pie_chart(top_parties_df, party)
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

# Function to create a subplot with pie charts for both years
def create_pie_chart_comparison(data, party):
    # Filter data for the specified party and each year
    data_2019 = data[(data['party'] == party) & (data['year'] == 2019)]['education'].value_counts(normalize=True) * 100
    data_2024 = data[(data['party'] == party) & (data['year'] == 2024)]['education'].value_counts(normalize=True) * 100
    
    # Create subplots for the two years
    fig = make_subplots(
        rows=1, cols=2, 
        specs=[[{'type': 'domain'}, {'type': 'domain'}]], 
        subplot_titles=[f'{party} (2019)', f'{party} (2024)']
    )
    
    # Add pie chart for 2019
    fig.add_trace(
        go.Pie(
            labels=data_2019.index,
            values=data_2019.values,
            name='2019',
            hole=0.3,
            textinfo='percent+label'
        ),
        row=1, col=1
    )
    
    # Add pie chart for 2024
    fig.add_trace(
        go.Pie(
            labels=data_2024.index,
            values=data_2024.values,
            name='2024',
            hole=0.3,
            textinfo='percent+label'
        ),
        row=1, col=2
    )
    
    # Update the layout
    fig.update_layout(
        title_text=f'Education Background Distribution - {party}',
        title_x=0.5,  # Center the title
        showlegend=False,  # Hide legends for better clarity
    )
    
    # Show the figure
    fig.show()

# Generate the pie chart comparison for all parties
parties = top_parties_df['party'].unique()
for party in parties:
    create_pie_chart_comparison(top_parties_df, party)
```

### 6. **Outlier Detection**

Identify outliers in `age`, `total_assets`, `liabilities`, and `net_worth` for each year separately.

```python
import pandas as pd
import numpy as np
import plotly.express as px

# Load the dataset (assuming `top_parties_df` is the DataFrame with relevant data)
# Replace 'total_assets' and 'criminal_cases' with the column names for net worth and criminal cases in your dataset.
top_parties_df['total_assets'] = top_parties_df['total_assets'].replace('Not Given', np.nan).astype(float)
top_parties_df['criminal_cases'] = top_parties_df['criminal_cases'].replace('Not Given', np.nan).astype(float)

# Define a function to identify outliers using the IQR method
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)  # First quartile (25th percentile)
    Q3 = data[column].quantile(0.75)  # Third quartile (75th percentile)
    IQR = Q3 - Q1  # Interquartile range
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return (data[column] < lower_bound) | (data[column] > upper_bound)

# Detect outliers for 'total_assets' (net worth) and 'criminal_cases'
top_parties_df['outlier_assets'] = detect_outliers_iqr(top_parties_df, 'total_assets')
top_parties_df['outlier_criminal_cases'] = detect_outliers_iqr(top_parties_df, 'criminal_cases')

# Create a new column to identify combined outliers (either net worth or criminal cases)
top_parties_df['outlier'] = top_parties_df['outlier_assets'] | top_parties_df['outlier_criminal_cases']

# Scatter plot with Plotly Express
fig = px.scatter(
    top_parties_df,
    x='total_assets',
    y='criminal_cases',
    color='outlier',
    hover_data=['candidate_name', 'party', 'state'],  # Include details in hover
    labels={
        'total_assets': 'Net Worth (in Crores)',
        'criminal_cases': 'Criminal Cases'
    },
    title="Scatter Plot of Net Worth vs Criminal Cases with Outliers Highlighted",
    color_discrete_map={True: 'red', False: 'blue'}  # Red for outliers, Blue for normal
)

# Update layout for better visualization
fig.update_traces(marker=dict(size=10, line=dict(width=1, color='black')))
fig.update_layout(
    xaxis_title="Net Worth (in Crores)",
    yaxis_title="Criminal Cases",
    legend_title="Outlier",
    title_x=0.5
)

# Show the plot
fig.show()
```
